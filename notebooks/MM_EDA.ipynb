```python
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from tensorflow.keras.layers import Dense
import tensorflow as tf

```

Baseline Neural Network Model Definition

```python
# define base model
def baseline_model():
	# create model
	model = Sequential()
	model.add(Dense(13, input_dim=5, kernel_initializer='normal', activation='relu'))
	model.add(Dense(1, kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_absolute_error', optimizer='adam')
	return model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import numpy as np

y_pred = baseline_model().predict(X_train)
mae = mean_absolute_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
print(f'Train Mean Absolute Error: {mae:.2f}')
print(f'Train R^2: {r2:.2f}')
print(f'Train Mean Squared Error: {mse:.2f}')
print(f'Train Root Mean Squared Error: {rmse:.2f}')
```

```python
y_pred = baseline_model().predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Test Mean Absolute Error: {mae:.2f}')
print(f'Test R^2: {r2:.2f}')
print(f'Test Mean Squared Error: {mse:.2f}')
print(f'Test Root Mean Squared Error: {rmse:.2f}')
```

The Multi-Layer Neural Network Model

```python
from tensorflow.keras.layers import LSTM
# The Model
def create_model(lr:float=0.01):
  opt = tf.keras.optimizers.Adam(
    learning_rate=lr, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False,
    name='Adam')

  model = Sequential([
          Dense(13, input_dim=5, kernel_initializer='normal', activation='relu'), # The Input Layer
          Dense(256, kernel_initializer='normal',activation='relu'), # Hidden Layer
          Dense(256, kernel_initializer='normal',activation='relu'), # Hidden Layer
          Dense(1, kernel_initializer='normal',activation='linear') #The Output Layer

  ])

  model.compile(optimizer=opt, loss='mean_absolute_error',
                    metrics=['mean_absolute_error'])
  return model
```

```python
y_pred = create_model().predict(X_train)
mae = mean_absolute_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
print(f'Train Mean Absolute Error: {mae:.2f}')
print(f'Train R^2: {r2:.2f}')
print(f'Train Mean Squared Error: {mse:.2f}')
print(f'Train Root Mean Squared Error: {rmse:.2f}')
```

```python
y_pred = create_model().predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Test Mean Absolute Error: {mae:.2f}')
print(f'Test R^2: {r2:.2f}')
print(f'Test Mean Squared Error: {mse:.2f}')
print(f'Test Root Mean Squared Error: {rmse:.2f}')
```
